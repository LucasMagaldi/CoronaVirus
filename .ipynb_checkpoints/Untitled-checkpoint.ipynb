# Comparando dados do Covid por estados

```python
import pandas as pd
import numpy as np
import fbprophet as fbp
from fbprophet import Prophet
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/sandeco/CanalSandeco/master/covid-19/covid_19_data.csv")
```

```python
df.head()
```

```python
df.rename(columns={'ObservationDate': 'Date', 'Country/Region': 'Country'}, inplace = True)
```

```python
df.head()
```

# Ajustando por data

```python
mortes = df.groupby('Date').sum()['Deaths'].reset_index()
mortes.tail()
```

```python
mortes.plot()
```

# Previsão de mortes COVID - 19 no mundo utilizando prophet

```python
mortes.columns = ['ds', 'y']
mortes.head()
```

```python
mortes['ds'] = pd.to_datetime(mortes['ds'])
mortes.tail()
```

```python
dias = 30
```

```python
m = Prophet(interval_width=0.95)
m.fit(mortes)
futuro = m.make_future_dataframe(periods= dias)
```

```python
futuro.tail(dias)
```

# Iniciando a previsão 

```python
previsao = m.predict(futuro)
previsao.tail(dias)
```

```python
previsao[['ds', 'yhat_lower', 'yhat', 'yhat_upper']].tail(dias)
```

```python
mortes.tail(dias)
```

```python
confirmed_forecast_plot = m.plot(previsao)
```
